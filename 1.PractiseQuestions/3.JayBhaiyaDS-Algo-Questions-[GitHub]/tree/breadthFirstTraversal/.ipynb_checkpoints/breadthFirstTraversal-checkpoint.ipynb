# Breadth First Traversal



```javascript
class tree{
    constructor(value, left = null, right = null ){
        this.value = value;
        this.left = left;
        this.right = right;
    }
}


let a = new tree(1);
let b = new tree(2);
let c = new tree(3);
let d = new tree(4);
let e = new tree(5);
let f = new tree(6);
let g = new tree(7);
let h = new tree(8);
let i = new tree(9);
let j = new tree(10);
let k = new tree(11);

a.left = b;
b.left = d;
c.left = f;
d.left = i;
e.left = h;

a.right =  c;
b.right =  e;
c.right =  g;
d.right =  j;
e.right =  k;

class queue{
    constructor(){
        this.item = [];
    }
    enqueue(data){
        this.item = data;
    }
    dequeue(){
        this.item.pop();
    }
}

function breadthFirstTraversal(tree){
    let node = tree;
    let output = [];

    while(node){
        queue.enqueue(node.value);
        if(node.left){
	        queue.engueue(node.left);
        }
       	if(node.right){
	       	queue.engueue(node.right);
       	} 
        let element = queue.dequeue();
        output.push(element);
        node = node.left;
    }
}
```
